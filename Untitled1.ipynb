```python
import numpy as np
import pandas as pd
import re
import tqdm as tqdm
```

```python
def convert_kaggle(dataframe,dict_mapping):
    list_label=[]
    maximum=np.max(list(dict_mapping.values()))
    sum_total=np.sum(list(dict_mapping.values()))
    for index,row in dataframe.iterrows():
        if(row[2:].sum()==0):
            print(row[2:])
            list_label.append(0)
        else:
            final_label=0
            count=0
            for col in list(dataframe.columns)[2:]:
                print('rowcol',row[col])
                if(row[col]==1):
                    final_label+=dict_mapping[col]
                    count+=1
            if(final_label==0):
                final_label=sum_total/len(dict_mapping)
            else:
                final_label=final_label/count
            list_label.append(final_label/maximum)
    return list_label
```

```python
def get_dictionary(params):
    
    if(params['dataset']=='toxic_comment'):
        dict_mapping={
            'severe_toxic':9,
            'obscene':5,
            'threat':8,
            'insult':6,
            'identity_hate':7,
#             'racial':2,
            'toxic':4
        }
        
    elif(params['dataset']=='toxic_unintended'):
        dict_mapping={
            'severe_toxic':9,
            'obscene':5,
            'threat':8,
            'insult':6,
            'identity_hate':7
        }
    
    return dict_mapping
```

```python
df_train = pd.read_csv("Dataset/jigsaw-toxic-comment-classification-challenge/train.csv")
df_test = pd.read_csv("Dataset/jigsaw-toxic-comment-classification-challenge/test.csv")
df_test_labels = pd.read_csv("Dataset/jigsaw-toxic-comment-classification-challenge/test_labels.csv")
df_test=df_test.merge(df_test_labels, how='inner', on='id')
df_total=pd.concat([df_train,df_test]).reset_index(drop=True)
df_total=df_total[df_total['toxic']!=-1]

# ### uncomment to increase score for racial hate
# racial_list = open('Dataset/HateLexicons.txt').read().splitlines()
# racial_list = [re.escape(s) for s in racial_list]
# pattern = '|'.join(racial_list)
# df_total['racial'] = df_total['comment_text'].str.match(pattern, case=False)
# df_total['racial'] = df_total['racial'].astype(int)
# change dict mapping and convert_kaggle accordingly

# df_total = df_total[df_total.toxic == 1].head(5)

params = {'dataset':'toxic_comment'}
dict_mapping=get_dictionary(params)

list_labels=convert_kaggle(df_total,dict_mapping)
df_total['label']=list_labels

df_total = df_total.rename(columns={"comment_text": "text"})

```
